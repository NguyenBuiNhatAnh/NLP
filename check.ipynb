```python
import pandas as pd

# =========================
# LOAD FILE
# =========================
df = pd.read_csv("test_predictions.csv")

# =========================
# SHOW COLUMNS
# =========================
print("Columns:")
print(df.columns.tolist())

# =========================
# CHỈNH TÊN CỘT Ở ĐÂY
# =========================
TEXT_COL = "Nội dung"
TRUE_COL = "Nhãn thực tế"
PRED_COL = "Nhãn dự đoán"

# =========================
# FIND WRONG PREDICTIONS
# =========================
wrong_df = df[df[TRUE_COL] != df[PRED_COL]].copy()

print("\nWrong predictions:", len(wrong_df))

display(
    wrong_df[[TEXT_COL, TRUE_COL, PRED_COL]].head(20)
)

# =========================
# SAVE
# =========================
wrong_df.to_csv(
    "wrong_predictions.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\nSaved wrong_predictions.csv")
```
